```python
data['Age']=data['Age'].map({'<=30':1,'31...40':2,'31..40':2,'>40':3})
data['Income']=data['Income'].map({'high':1,'medium':2,'low':3})
data['Student']=data['Student'].map({'yes':1,'no':2})
data['Credit']=data['Credit'].map({'fair':0,'excellent':1})
data['Buys']=data['Buys'].map({'yes':1,'no':0})
```

```python
m=data.values
print(m)
```

```python
#Calculating probabilities
P=[]
yes=0
no=0
for i in range(0,len(m)):
    if(m[i][4]==1):
        yes=yes+1
    if(m[i][4]==0):
        no=no+1
P.append(round(yes/len(m),3))
P.append(round(no/len(m),3))
print(yes)
print(no)
print(P)
```

```python
#Predicting with respect to given conditions
#age <=30

n=0
y=0
for i in range(0,len(m)):
    if (m[i][0]==1 and m[i][4]==1):
        y+=1
    elif (m[i][0]==1 and m[i][4]==0):
        n+=1
print(y)
print(n)
P.append(round(y/yes,2))

P.append(round(n/no,2))
print(P)
```

```python
#income=medium
n=0
y=0
k=yes
e=no
for i in range(0,len(m)):
    if(m[i][1]==2 and m[i][4]==1):
        y=y+1
    if(m[i][1]==2 and m[i][4]==0):
        n=n+1
P.append(round(y/yes,2))
P.append(round(n/no,2))
print('Medium income buying computer',y)
print('Medium income not buying computer',n)
print(P)
```

```python
#student 
n=0
y=0
for i in range(0,len(m)):
    if(m[i][2]==1 and m[i][4]==1):
        y=y+1
    if(m[i][2]==1 and m[i][4]==0):
        n=n+1
P.append(round(y/yes,2))
P.append(round(n/no,2))
print('Student buying computer',y)
print('Student not buying computer',n)
```

```python
#Credit rating is fair
n=0
y=0
for i in range(0,len(m)):
    if(m[i][3]==0 and m[i][4]==1):
        y=y+1
    if(m[i][3]==0 and m[i][4]==0):
        n=n+1
P.append(round(y/yes,2))
P.append(round(n/no,2))
print('Credit rating fair buys computer',y)
print('Credit rating does not buy computer',n)
print(P)
```

```python
#Calculating final probabilities
F=[]
print(P)
print('-------------------------------------------')
y=1
n=1
for i in range(0,len(P)):
    if(i%2==0):
        y=y*P[i]
    else:
        n=n*P[i]
print(y)
print(n)
print('------------------------------------------')
F.append(y)
F.append(n)
print("The probability of Buying a computer")
for i in range(0,len(P)):
    if(i%2==0):
        print(P[i])
print("The probability of not Buying a computer")
for i in range(0,len(P)):
    if(i%2!=0):
        print(P[i])
print("NAIVE BAYES for buying the computer")
print(F[0])
print("NAIVE BAYES for not buying the computer")
print(F[1])
if (F[0]>F[1]):
    print('User buys computer')
else:
    print('User does not buy computer')

```
